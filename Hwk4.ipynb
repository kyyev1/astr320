```python
#import libraries
import pandas as pd

import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)

#read photometry
photoB = pd.read_csv ('Bpic.dat', sep='\s+', header=None, names = ['ID','mag','error','fits'])
photoV = pd.read_csv ('Vpic.dat', sep='\s+', header=None, names = ['ID','mag','error','fits'])

#print measured magnitudes
print(photoB)
print(photoV)

#make arrays for magnitudes of photoB and photoV
arrayB=photoB['mag']
arrayV=photoV['mag']

#make array for B-V
BV = arrayB - arrayV

#plot V vs. V-B
plt.plot(BV, arrayV, 'ro')
plt.xlabel('B-V')
plt.ylabel('V')
plt.gca().invert_yaxis()
plt.title('M45 Color Magnitude Diagram')
plt.show()

#should have found magnitudes for more stars for a more complete diagram
```
